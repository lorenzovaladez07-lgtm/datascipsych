## Problem (graduate students): conditionals (4 points)

Read about [conditionals](https://docs.pola.rs/user-guide/expressions/basic-operations/#conditionals), which allow you to define values in a column that vary based on conditions. The values may be specific values (e.g., numbers or strings) or may be taken from a column in the data frame. To indicate a literal string value, use `pl.lit` so Polars will know that you mean to refer to a string rather than a column name.

For example, if we have columns `trial_type` (`"target"` or `"lure"`) and `response` (`"old"` or `"new"`), then this code will create a `response_type` column indicating whether each trial was a hit, miss, false alarm, or correct rejection. If some unexpected combination of trial type and response is encountered, it will be assigned a `null` value.

```python
df.with_columns(
    response_type=pl.when((pl.col("trial_type") == "target") & (pl.col("response") == "old"))
    .then(pl.lit("hit"))
    .when((pl.col("trial_type") == "target") & (pl.col("response") == "new"))
    .then(pl.lit("miss"))
    .when((pl.col("trial_type") == "lure") & (pl.col("response") == "old"))
    .then(pl.lit("false alarm"))
    .when((pl.col("trial_type") == "lure") & (pl.col("response") == "new"))
    .then(pl.lit("correct rejection"))
    .otherwise(None)
)
```

Note that, just like Python's `if/elif/else` statements, we can include as many `when/then` pairs as we want to test different conditions and assign a value for each one, and we can optionally have an `otherwise` expression to indicate the value to use if none of the `when` expressions are `True`.
